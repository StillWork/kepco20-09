# groupby  연습

- 각 그룹에서 하나의 값을 얻을 때: aggregate()
 - mean, sum, size, 사용자 정의 함수
- 원래 행에서 일부 서브셋을 얻을 때: filter()
 - 함수를 적용하여 True/False를 얻는다
- 각 행에 대해서 새로운 값을 얻을 때: transform()

# 데이터
- 식당 이용 데이터

```python
import pandas as pd
df = pd.read_csv(
   "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
   )
# pd.options.display.max_rows = 10
df
```

```python
df.day.unique()
```

```python
df.day.value_counts().to_frame()
```

# 연습

## 요일별 total_bill의 평균은?

```python
#  과제 수행 ...  agg() 사용

df.groupby('day')['total_bill'].mean().to_frame()
```

```python
df.groupby('day')['total_bill', 'tip'].agg(['mean'])
```

```python
# 참고
df.groupby('day')['total_bill'].mean()
```

```python
# 전체 보기
df.groupby('day').agg('mean')
```

## 평균 요금이 20 을 넘는 날(요일)의 데이터를 추출하시오
- 토, 일 해당

```python
## filter() 사용
```

```python
df.groupby('day').filter(
    lambda x : x['total_bill'].mean() > 20 # 결과는 불리언
    )
```

```python
df.total_bill > 20
```

```python
# 주의: 아래의 결과와 다름!
# 아래는 각 샘플중에 요금이 20달러 이상인 날을 찾음 (97일)
df[df.total_bill > 20]
```

## 평균이 20을 넘는 날 (토, 일)의 평균 손님 수는?

```python
## size의 평균을 구해야 함











```

```python
df.groupby('day').filter(
    lambda x : x['total_bill'].mean() > 20
)['size'].mean()
```

## 평균이 20이 안되는 요일의 손님 수 평균은?

```python
##

```

```python
df.groupby('day').filter(
    lambda x : x['total_bill'].mean() <= 20
)[['size','tip']].mean()
```

## 요일별 평균 매출액 컬럼을 추가하시오

```python
## 새로운 값을 얻으므로 transform()사용
## 컬럼명 'day_average' 추가

```

```python
df['day_average'] = df.groupby('day')['total_bill'].transform(
    lambda x : x.mean()
)
df
```

## 평균 매출로 정규화한 일일 매출액을 구하시오

```python
##  'total_bill'을 'day_average'로 나누어주면 됨












```

```python
# 한가지 방법
df['ratio'] = df['total_bill'] / df['day_average']
```

```python
df
```

## 위의 과정을 transform()을 이용하여 한번에 구하시오

```python
##










```

```python
# 한번에 구하는 방법
df['scaled bill'] = df.groupby('day')['total_bill'].transform(
    lambda x : x/x.mean()
)
df.head()
```

# 참고
- groupby()의 결과는 그룹명과 데이터프레임 두부분을 리턴한다
- 아래와 같은 iteration이 가능하다

```python
df.groupby('day').get_group('Sat')
```

```python
for day, group in df.groupby('day'):
    day_mean = group['total_bill'].mean()
    print('average bill for ' + day + ' is ' + str(day_mean))
```

```python
# 활용 방안 (각각 별도의 파일로 저장하기)
for day, group in df.groupby('day'):
    group.to_csv(day + '.csv')
```
